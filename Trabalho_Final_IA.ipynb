```python
df = df.rename(columns={'name': 'Nome',
    'pokedex id': 'Pokedex ID',
    'height': 'Altura',
    'weight': 'Peso',
    'type': 'Tipo',
    'secundary type': 'Tipo Secundário',
    'hp': 'Vida',
    'attack': 'Ataque',
    'defense': 'Defesa',
    'sp atk': 'Ataque especial',
    'sp def': 'Defesa especial',
    'speed': 'Velocidade'                       #TRASNFORMANDO o dataframe, no caso trocando o nome das colunas para melhor entendimento
})

df.head(1000)
```

```python
#Dataframe com os 15 pokemons que faltavam para completar a 8 geração
novos_pokemons = pd.DataFrame({
    'Nome': ['Kubfu', 'Urshifu', 'Zarude', 'Regieleki', 'Regidrago', 'Glastrier', 'Spectrier', 'Calyrex', 'Wyerdeer', 'Kleavor', 'Ursaluna', 'Basculegion', 'Sneasler', 'Overqwil', 'Enamorus'],
    'Pokedex ID': [891, 892, 893, 894, 895, 896, 897, 898, 899, 900, 901, 902, 903, 904, 905],
    'Altura': [6, 19, 17, 12, 21, 22, 20, 1.1, 1.7, 0.4, 1.2, 3.6, 2.1, 1.5, 0.6],
    'Peso': [12, 105, 700, 145, 2000, 800, 44.5, 7.7, 78.2, 9.9, 75, 260, 28, 10, 1.8],
    'Tipo': ['fighting', 'fighting', 'dark', 'electric', 'dragon', 'ice', 'ghost', 'psychic', 'normal', 'bug', 'psychic', 'water', 'dark', 'dark', 'bug'],
    'Tipo Secundário': ['None', 'dark', 'grass', 'None', 'None', 'None', 'None', 'grass', 'None', 'flying', 'fairy', 'water', 'ice', 'ghost', 'psychic'],
    'Vida': [60, 100, 105, 80, 200, 100, 100, 100, 100, 62, 107, 100, 70, 80, 92],
    'Ataque': [90, 130, 120, 100, 164, 145, 65, 80, 110, 150, 131, 123, 95, 125, 65],
    'Defesa': [60, 100, 105, 60, 90, 130, 60, 80, 105, 60, 97, 100, 65, 60, 45],
    'Ataque especial': [53, 63, 70, 100, 100, 65, 60, 80, 80, 75, 97, 120, 45, 50, 95],
    'Defesa especial': [50, 60, 95, 80, 100, 110, 110, 80, 110, 110, 97, 100, 45, 100, 110],
    'Velocidade': [72, 97, 105, 200, 80, 30, 130, 80, 60, 145, 97, 121, 75, 90, 60]
})

novos_pokemons['Altura'] = novos_pokemons['Altura'].round() .astype(int)
novos_pokemons['Peso'] = novos_pokemons['Peso'].round().astype(int)

df = novos_pokemons.append(df, ignore_index=True)
```

```python
primeiras_linhas = df.head(15)

df = df.drop(primeiras_linhas.index)      #Tirando os 15 primeiros (novos pokemons) que eu adicionei no começo do df e colocando no final

df = pd.concat([df, primeiras_linhas])
```

```python
df.set_index('Pokedex ID', inplace=True) #Trocando o index comum pelo Pokemon ID de index

df.head(1000) 
```

```python
tipos_unicos = df['Tipo'].unique()
                                     #Vendo os valores únicos da coluna Tipo pra poder trocar os nomes deles para melhor visualização
print(tipos_unicos)
```

```python
trocasdetipos = {
        'grass': 'Grama',
    'fire': 'Fogo',            #Criando o dicionário para dar de argumento no replace pq o replace so recebe até 7 argumentos
    'water': 'Água',
    'bug': 'Inseto',
    'poison': 'Venenoso',
    'electric': 'Elétrico',
    'ground': 'Terra',
    'fairy': 'Fada',
    'fighting': 'Lutador',
    'psychic': 'Psíquico',
    'rock': 'Pedra',
    'ghost': 'Fantasma',
    'ice': 'Gelo',          
    'dragon': 'Dragão',
    'dark': 'Sombrio',
    'flying': 'Voador',
    'steel' : 'Aço'
}

df['Tipo'] = df['Tipo'].replace (trocasdetipos)
df['Tipo Secundário'] = df['Tipo Secundário'].replace (trocasdetipos)      #trocando os nomes dos tipos

df.head(10000)
```

```python
#OBSERVAÇÕES CATEGORICAS
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot()
plt.title('Correlação entre HP e Peso')

ax.scatter(df['Vida'], df['Peso'])  #setando as colunas usadas

ax.set_xlabel('Vida')                     #grafico
ax.set_ylabel('Peso')


x = df['Vida']
y = df['Peso']
fit = np.polyfit(x, y, 2)              #linha de tendência
fit_fn = np.poly1d(fit)
plt.plot(x, fit_fn(x), color='red')


plt.show()        

# Usando a curva de tendêndo podemos provar a ideia de que quanto mais pesado mais Vida tem o Pokemon e vice versa
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.title('Correlação entre Velocidade e Peso')

ax.scatter(df['Velocidade'], df['Peso'])  #setando as colunas usadas

ax.set_xlabel('Velocidade')                     #grafico
ax.set_ylabel('Peso')


x = df['Velocidade']
y = df['Peso']
fit = np.polyfit(x, y, 2)              #linha de tendência
fit_fn = np.poly1d(fit)
plt.plot(x, fit_fn(x), color='red')


plt.show()     

# Já correlacionando a Velocidade e o Peso podemos evidenciar que eles não tem muito a ver um com o outro
# pois a linha de tendência se mantem reta e não crescente
```
